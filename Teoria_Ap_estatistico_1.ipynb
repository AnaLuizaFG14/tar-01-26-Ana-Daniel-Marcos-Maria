```r

dados <- read.csv("Santos2018_2025.csv")

tempo_num <- as.numeric(dados$tempo_atracado)


media <- mean(tempo_num)
mediana <- median(tempo_num)
moda <- as.numeric(names(tempo_num)[tempo_num == max(tempo_num)])
variancia <- var(tempo_num)
desvio_padrao <- sd(tempo_num)
mediana_geometrica <- exp(mean(log(tempo_num)))
desvio_absoluto <- mean(abs(tempo_num - mean(tempo_num)))
MAD <- mad(tempo_num, constant = 1)
amplitude <- diff(range(tempo_num))


# Mostrar resultados
paste("Média = ", media)
paste("Mediana = ", mediana)
paste("Variância = ", variancia)
paste("Desvio Padrão = ", desvio_padrao)
paste("Mediana Geometrica = ", mediana_geometrica)
paste("Desvio Absoluto Médio = ", desvio_absoluto)
paste("MAD = ", MAD)
paste("Amplitude = ", amplitude)
paste("1 Quartil = ", quantile(tempo_num, probs = 0.25))
paste("2 Quartil = ", quantile(tempo_num, probs = 0.50))
paste("3 Quartil = ", quantile(tempo_num, probs = 0.75))
paste("4 Quartil = ", quantile(tempo_num, probs = 1))
paste("Amplitude Interquartilica = ", IQR(tempo_num))
```
